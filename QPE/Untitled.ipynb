```python
from qiskit import QuantumCircuit

# Create a circuit with 4 qubits (2 data + 2 register qubits)
qc = QuantumCircuit(4)

# Step 1: Create superposition and entanglement
qc.h(0)
qc.cx(0, 1)

# Step 2: Map register states
qc.ccx(0, 1, 2)  # If |11> set register qubit to |1>
qc.cx(1, 3)      # If |01> set second register qubit to |1>

# Draw the circuit
qc.draw("mpl")
```

```python
from qiskit import QuantumCircuit
import numpy as np

# Create a circuit with 4 qubits (2 data + 2 register qubits)
qc = QuantumCircuit(4)

# Step 1: Initialize data qubits
qc.h(0)           # Create superposition on first qubit
qc.cx(0, 1)       # Entangle the second qubit

# Step 2: Controlled rotations for amplitude preservation
qc.cry(np.pi/4, 0, 2)  # Controlled rotation from qubit 0 to register qubit 2
qc.cry(np.pi/2, 1, 3)  # Controlled rotation from qubit 1 to register qubit 3

# Draw the circuit
qc.draw("mpl")
```
